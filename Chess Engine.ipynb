```python
import tkinter as tk
from tkinter import ttk
import numpy as np
import copy
import random
import math

#https://github.com/aqeelanwar/Snake-And-Apple/blob/master/main.py


size_of_board = 8

space_size = 50



class Chess():
    
    #-----------------------------------------------------------------------------------------
    #Initialization Functions
    #-----------------------------------------------------------------------------------------
    def __init__(self):
        

        self.window = tk.Tk()
        self.window.title("Chess")
        self.MainFrame = ttk.Frame(self.window)
        
        self.playArea = tk.Canvas(self.MainFrame, height = size_of_board*space_size, width = size_of_board*space_size, bg = "gray")
        self.playArea.grid(row = 1,column = 0)
        
        #self.newGameButton = ttk.Button(self.MainFrame, text = \"New Game\", command = self.newGame())
        #self.newGameButton.grid(row = 0,column = 0)

        #self.QuitButton = ttk.Button(self.MainFrame, text=\"Quit\", command = self.window.destroy)
        #self.QuitButton.grid(row = 2, column=0)
        
        self.MainFrame.grid()
        
        #Input from User
        self.window.bind('<space>', self.newGame)
        self.window.bind('<Button-1>', self.click)
        
        self.gameover = False
        
        
        self.whiteAttackFlag = False
        self.blackAttackFlag = False
        
        self.blackKingInCheck = False
        self.whiteKingInCheck = False
        
        self.player = "White"
        
        self.turn = "White"
        
        self.pieces = ['NA','WPawn','WRook','WKnight','WBishop','WQueen','WKing','KPawn','KRook','KKnight','KBishop','KQueen','KKing']
        
        self.boardPieces = []
        self.checkMoves = []
        
        self.newGame()
        
        
        
    def newGame(self):
        
        self.initializeBoard()
        self.firstClick = True
        print(self.turn)
        
        
    def mainloop(self):
        self.window.mainloop()
                
    #-----------------------------------------------------------------------------------------------------------------------------------------------
    #Drawing Functions
    #The Modules required to draw required game based object on canvas
    #-----------------------------------------------------------------------------------------------------------------------------------------------
    
    def initializeBoard(self):
        self.playArea.delete("all")
        self.gameBoard = np.zeros([size_of_board,size_of_board])
        self.pawnList = [[0,1],[1,1],[2,1],[3,1],[4,1],[5,1],[6,1],[7,1],[0,6],[1,6],[2,6],[3,6],[4,6],[5,6],[6,6],[7,6]]
        for i in range(size_of_board):
            
            self.playArea.create_line(i*space_size,
                            0,
                            i*space_size,
                            size_of_board*space_size, width = 1)
            self.playArea.create_line(0,
                            i*space_size,
                            size_of_board*space_size,
                            i*space_size, width = 1)
        
        
            if self.player == "White":
                self.whiteplus = 2.5
                self.blackplus = -2.5
            elif self.player == "Black":
                self.whiteplus = -2.5
                self.blackplus = 2.5
                
            self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.blackplus+4)*space_size, text="P", fill="black", font=('Helvetica 15 bold')))
            self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.whiteplus+4)*space_size, text="P", fill="white", font=('Helvetica 15 bold')))
            self.gameBoard[i,int(self.blackplus+3.5)] = 7
            self.gameBoard[i,int(self.whiteplus+3.5)] = 1
            
            if self.player == "White":
                self.whiteplus = 3.5
                self.blackplus = -3.5
            elif self.player == "Black":
                self.whiteplus = -3.5
                self.blackplus = 3.5
            
            if i in [0,7]:
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.blackplus+4)*space_size, text="R", fill="black", font=('Helvetica 15 bold')))
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.whiteplus+4)*space_size, text="R", fill="white", font=('Helvetica 15 bold')))
                self.gameBoard[i,int(self.blackplus+3.5)] = 8
                self.gameBoard[i,int(self.whiteplus+3.5)] = 2
            if i in [1,6]:
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.blackplus+4)*space_size, text="Kn", fill="black", font=('Helvetica 15 bold')))
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.whiteplus+4)*space_size, text="Kn", fill="white", font=('Helvetica 15 bold')))
                self.gameBoard[i,int(self.blackplus+3.5)] = 9
                self.gameBoard[i,int(self.whiteplus+3.5)] = 3
            if i in [2,5]:
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.blackplus+4)*space_size, text="B", fill="black", font=('Helvetica 15 bold')))
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.whiteplus+4)*space_size, text="B", fill="white", font=('Helvetica 15 bold')))
                self.gameBoard[i,int(self.blackplus+3.5)] = 10
                self.gameBoard[i,int(self.whiteplus+3.5)] = 4
            if i == 3:
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.blackplus+4)*space_size, text="Q", fill="black", font=('Helvetica 15 bold')))
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.whiteplus+4)*space_size, text="Q", fill="white", font=('Helvetica 15 bold')))
                self.gameBoard[i,int(self.blackplus+3.5)] = 11
                self.gameBoard[i,int(self.whiteplus+3.5)] = 5
            if i == 4:
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.blackplus+4)*space_size, text="K", fill="black", font=('Helvetica 15 bold')))
                self.boardPieces.append(self.playArea.create_text((i+0.5)*space_size, (self.whiteplus+4)*space_size, text="K", fill="white", font=('Helvetica 15 bold')))
                self.gameBoard[i,int(self.blackplus+3.5)] = 12
                self.gameBoard[i,int(self.whiteplus+3.5)] = 6
        
            
    def reDraw(self):
        for piece in self.boardPieces:
            self.playArea.delete(piece)
        self.boardPieces = []
        for col in range(size_of_board):
            for row in range(size_of_board):
                if self.gameBoard[col,row] == 1:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="P", fill="white", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 2:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="R", fill="white", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 3:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="Kn", fill="white", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 4:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="B", fill="white", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 5:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="Q", fill="white", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 6:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="K", fill="white", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 7:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="P", fill="black", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 8:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="R", fill="black", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 9:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="Kn", fill="black", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 10:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="B", fill="black", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 11:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="Q", fill="black", font=('Helvetica 15 bold')))
                elif self.gameBoard[col,row] == 12:
                    self.boardPieces.append(self.playArea.create_text((col+0.5)*space_size, (row+0.5)*space_size, text="K", fill="black", font=('Helvetica 15 bold')))
        
        
    def display_gameover(self):
        self.playArea.delete("all")
        self.playArea.create_text(
            size_of_board / 2,
            3 * size_of_board / 8,
            font="cmr 40 bold",
            fill="Green",
            text="Game Over",
        )
    def validMoveDraw(self):
        self.moveOptions = []
        for move in self.validMoves:
            self.moveOptions.append(self.playArea.create_rectangle(move[0]*space_size,
                                                                  move[1]*space_size,
                                                                  move[0]*space_size+space_size,
                                                                  move[1]*space_size+space_size,fill="green"))
        for move in self.moveOptions:
            self.playArea.tag_lower(move)
    #---------------------------------------------------------------------------------------------------------------------------------------------
    #Logical Function
    #The modules required to carry out the game function
    #---------------------------------------------------------------------------------------------------------------------------------------------
    def update_bird(self):
        
        self.gravity()
    
        self.updatePosition()
    
    def click(self, e):
        column = int(np.floor(e.x/space_size))
        row = int(np.floor(e.y/space_size))
        
        #print(self.pieces[int(self.gameBoard[column,row])])
        
        if self.firstClick:
            
            self.selectedPiece = [column,row]
            if (self.gameBoard[self.selectedPiece[0]][self.selectedPiece[1]] in [1,2,3,4,5,6] and self.turn == "White") or \
                (self.gameBoard[self.selectedPiece[0]][self.selectedPiece[1]] in [7,8,9,10,11,12] and self.turn == "Black"):
                
                self.firstSelect = self.playArea.create_rectangle(column*space_size,
                                                                  row*space_size,
                                                                  column*space_size+space_size,
                                                                  row*space_size+space_size,fill="blue")

                self.playArea.tag_lower(self.firstSelect)


                
                self.calculateValidMoves(self.selectedPiece)
                #Display Moves as green spaces-------------------------------------------------------------------------------------
                self.validMoveDraw()
                
                self.firstClick = False
            
        else:
            self.moveSpace = [column,row]
            
            if self.moveSpace in self.validMoves:
                
                tempBoard = copy.copy(self.gameBoard)
                
                interim = self.gameBoard[self.selectedPiece[0]][self.selectedPiece[1]]
                self.gameBoard[self.selectedPiece[0]][self.selectedPiece[1]] = 0
                self.gameBoard[self.moveSpace[0]][self.moveSpace[1]] = interim
                
                
                if ((self.turn == "White" and self.whiteKingInCheck) or (self.turn == "Black" and self.blackKingInCheck)) and np.any(np.all(self.gameBoard == self.checkMoves, axis=1)):
                    print("Cant move there cause it leaves you in check")
                    self.moveSpace = self.selectedPiece
                    if self.turn == "White":
                        self.whiteKingInCheck = False
                    elif self.turn == "Black":
                        self.blackKingInCheck = False
                
                self.gameBoard = copy.copy(tempBoard)
                
                interim = self.gameBoard[self.selectedPiece[0]][self.selectedPiece[1]]
                self.gameBoard[self.selectedPiece[0]][self.selectedPiece[1]] = 0
                self.gameBoard[self.moveSpace[0]][self.moveSpace[1]] = interim
                
                self.calcWhiteAttackMoves()
                self.calcBlackAttackMoves()
                self.myKingInCheck()
                
                
                
                self.playArea.delete(self.firstSelect)
                
                for move in self.moveOptions:
                    self.playArea.delete(move)
                    
                self.firstClick = True
                
                if self.moveSpace != self.selectedPiece:
                    if self.selectedPiece in self.pawnList:
                        self.pawnList.pop(self.pawnList.index(self.selectedPiece))
                    
                    if self.turn == 'White':
                        self.turn = 'Black'
                    else:
                        self.turn = 'White'
                    print(self.turn)
                    self.enemyKingInCheckmate()
                
                self.reDraw()
                
            
        #print(self.gameBoard)
        
    def calculateValidMoves(self, Piece):
        self.validMoves = []
        
        if self.player == "White":
                self.direction = 1
        elif self.player == "Black":
                self.direction = -1
        
        #Pawns for white
        if self.gameBoard[Piece[0]][Piece[1]] == 1:
            if self.gameBoard[Piece[0]][Piece[1]-(1*self.direction)] == 0:
                self.validMoves.append([Piece[0],Piece[1]-(1*self.direction)])
            if Piece[0]-1>=0:
                if self.gameBoard[Piece[0]-1][Piece[1]-(1*self.direction)] != 0 :
                    self.validMoves.append([Piece[0]-1,Piece[1]-(1*self.direction)])
            if Piece[0]+1<=7:
                if self.gameBoard[Piece[0]+1][Piece[1]-(1*self.direction)] != 0 :
                    self.validMoves.append([Piece[0]+1,Piece[1]-(1*self.direction)])
            if self.gameBoard[Piece[0]][Piece[1]-(2*self.direction)] == 0:
                if Piece in self.pawnList:
                    self.validMoves.append([Piece[0],Piece[1]-(2*self.direction)])
                    
                    
        
        #Pawns for black
        if self.gameBoard[Piece[0]][Piece[1]] == 7:
            if self.gameBoard[Piece[0]][Piece[1]+(1*self.direction)] == 0:
                self.validMoves.append([Piece[0],Piece[1]+(1*self.direction)])
            if Piece[0]-1>=0:
                if self.gameBoard[Piece[0]-1][Piece[1]+(1*self.direction)] != 0:
                    self.validMoves.append([Piece[0]-1,Piece[1]+(1*self.direction)])
            if Piece[0]+1<=7:
                if self.gameBoard[Piece[0]+1][Piece[1]+(1*self.direction)] != 0:
                    self.validMoves.append([Piece[0]+1,Piece[1]+(1*self.direction)])
            if self.gameBoard[Piece[0]][Piece[1]+(2*self.direction)] == 0:
                if Piece in self.pawnList:
                    self.validMoves.append([Piece[0],Piece[1]+(2*self.direction)])
                    
                    
        
        #Rooks and Queens----------------------------------------------------------------------------------------
        if self.gameBoard[Piece[0]][Piece[1]] in [2,8,5,11]:
            #down move
            hitPiece = False
            i = 1
            while not hitPiece:
                if Piece[1]+(i)<=7:
                    if self.gameBoard[Piece[0]][Piece[1]+(i)] == 0:
                        self.validMoves.append([Piece[0],Piece[1]+(i)])
                        i = i+1
                    else:
                        self.validMoves.append([Piece[0],Piece[1]+(i)])
                        hitPiece = True
                    
                else:
                    hitPiece = True
            #up move
            hitPiece = False
            i = 1
            while not hitPiece:
                if Piece[1]-(i)>=0:
                    if self.gameBoard[Piece[0]][Piece[1]-(i)] == 0:
                        self.validMoves.append([Piece[0],Piece[1]-(i)])
                        i = i+1
                    else:
                        self.validMoves.append([Piece[0],Piece[1]-(i)])
                        hitPiece = True
                    
                else:
                    hitPiece = True
            #right move
            hitPiece = False
            i = 1
            while not hitPiece:
                if Piece[0]+(i)<=7:
                    if self.gameBoard[Piece[0]+i][Piece[1]] == 0:
                        self.validMoves.append([Piece[0]+i,Piece[1]])
                        i = i+1
                    else:
                        self.validMoves.append([Piece[0]+i,Piece[1]])
                        hitPiece = True
                    
                else:
                    hitPiece = True
            #left move
            hitPiece = False
            i = 1
            while not hitPiece:
                if Piece[0]-i>=0:
                    if self.gameBoard[Piece[0]-i][Piece[1]] == 0:
                        self.validMoves.append([Piece[0]-i,Piece[1]])
                        i = i+1
                    else:
                        self.validMoves.append([Piece[0]-i,Piece[1]])
                        hitPiece = True
                    
                else:
                    hitPiece = True
                    
        #Knights--------------------------------------------------------------------------------------------------
        if self.gameBoard[Piece[0]][Piece[1]] in [3,9]:
             
            self.validMoves.append([Piece[0]+1,Piece[1]+2])
            self.validMoves.append([Piece[0]+1,Piece[1]-2])
            self.validMoves.append([Piece[0]-1,Piece[1]+2])
            self.validMoves.append([Piece[0]-1,Piece[1]-2])
            self.validMoves.append([Piece[0]+2,Piece[1]+1])
            self.validMoves.append([Piece[0]+2,Piece[1]-1])
            self.validMoves.append([Piece[0]-2,Piece[1]+1])
            self.validMoves.append([Piece[0]-2,Piece[1]-1])
            i = 0
            while i < len(self.validMoves):
                if self.validMoves[i][0]<0 or self.validMoves[i][0]>7 or self.validMoves[i][1]<0 or self.validMoves[i][1]>7:
                    self.validMoves.pop(i)
                else:
                    i=i+1
            
            
        #Bishops and Queens--------------------------------------------------------------------------------------------------
        if self.gameBoard[Piece[0]][Piece[1]] in [4,10,5,11]:
            #down right move
            hitPiece = False
            i = 1
            while not hitPiece:
                if Piece[0]+i<=7 and Piece[1]+i<=7:
                    if self.gameBoard[Piece[0]+i][Piece[1]+i] == 0:
                        self.validMoves.append([Piece[0]+i,Piece[1]+i])
                        i = i+1
                    else:
                        self.validMoves.append([Piece[0]+i,Piece[1]+i])
                        hitPiece = True
                    
                else:
                    hitPiece = True
            #up right move
            hitPiece = False
            i = 1
            while not hitPiece:
                if Piece[0]+i<=7 and Piece[1]-i>=0:
                    if self.gameBoard[Piece[0]+i][Piece[1]-i] == 0:
                        self.validMoves.append([Piece[0]+i,Piece[1]-i])
                        i = i+1
                    else:
                        self.validMoves.append([Piece[0]+i,Piece[1]-i])
                        hitPiece = True
                    
                else:
                    hitPiece = True
            #down left move
            hitPiece = False
            i = 1
            while not hitPiece:
                if Piece[0]-(i)>=0 and Piece[1]+(i)<=7:
                    if self.gameBoard[Piece[0]-i][Piece[1]+i] == 0:
                        self.validMoves.append([Piece[0]-i,Piece[1]+i])
                        i = i+1
                    else:
                        self.validMoves.append([Piece[0]-i,Piece[1]+i])
                        hitPiece = True
                    
                else:
                    hitPiece = True
            #up left move
            hitPiece = False
            i = 1
            while not hitPiece:
                if Piece[0]-i>=0 and Piece[1]-(i)>=0:
                    if self.gameBoard[Piece[0]-i][Piece[1]-i] == 0:
                        self.validMoves.append([Piece[0]-i,Piece[1]-i])
                        i = i+1
                    else:
                        self.validMoves.append([Piece[0]-i,Piece[1]-i])
                        hitPiece = True
                    
                else:
                    hitPiece = True
                    
        #Kings-------------------------------------------------------------------------------------------------------------
        if self.gameBoard[Piece[0]][Piece[1]] in [6,12]:
            
                       
            
            self.validMoves.append([Piece[0]+1,Piece[1]-1])
            self.validMoves.append([Piece[0]+1,Piece[1]])
            self.validMoves.append([Piece[0]+1,Piece[1]+1])
            self.validMoves.append([Piece[0],Piece[1]-1])
            self.validMoves.append([Piece[0],Piece[1]+1])
            self.validMoves.append([Piece[0]-1,Piece[1]-1])
            self.validMoves.append([Piece[0]-1,Piece[1]])
            self.validMoves.append([Piece[0]-1,Piece[1]+1])
            i = 0
            while i < len(self.validMoves):
                if self.validMoves[i][0]<0 or self.validMoves[i][0]>7 or self.validMoves[i][1]<0 or self.validMoves[i][1]>7:
                    self.validMoves.pop(i)
                else:
                    i=i+1
        
        #Ignore moves that land on ally pieces-----------------------------------------------------------------------------
        
        i = 0
        while i < len(self.validMoves):
            if self.gameBoard[self.validMoves[i][0],self.validMoves[i][1]] in [1,2,3,4,5,6] and (self.turn == 'White' or self.whiteAttackFlag):
                self.validMoves.pop(i)
            elif self.gameBoard[self.validMoves[i][0],self.validMoves[i][1]] in [7,8,9,10,11,12] and (self.turn == 'Black' or self.blackAttackFlag):
                self.validMoves.pop(i)
            else:
                i=i+1

        self.validMoves.append([Piece[0],Piece[1]])
        

    def calcWhiteAttackMoves(self):
        self.whiteMoves = []
        self.whiteAttackFlag = True
        for col in range(size_of_board):
            for row in range(size_of_board):
                if self.gameBoard[col,row] <= 6 and self.gameBoard[col,row] != 0:
                    self.calculateValidMoves([col,row])
                    
                    for move in self.validMoves:
                        if move != [col,row]:
                            self.whiteMoves.append(move)
        temp = []
        for move in self.whiteMoves:
            if move not in temp:
                temp.append(move)
        self.whiteMoves = temp
        self.whiteAttackFlag = False
                    
    def calcBlackAttackMoves(self):
        self.blackMoves = []
        self.blackAttackFlag = True
        for col in range(size_of_board):
            for row in range(size_of_board):
                if self.gameBoard[col,row] > 6:
                    self.calculateValidMoves([col,row])
                    
                    for move in self.validMoves:
                        self.blackMoves.append(move)
        temp = []
        for move in self.blackMoves:
            if move not in temp:
                temp.append(move)
        self.blackMoves = temp
        self.blackAttackFlag = False
        
    def myKingInCheck(self):
        if [np.where(self.gameBoard == 6)[0][0], np.where(self.gameBoard == 6)[1][0]] in self.blackMoves:
            print("White King In Check")
            self.whiteKingInCheck = True
        else:
            self.whiteKingInCheck = False
            
        if [np.where(self.gameBoard == 12)[0][0], np.where(self.gameBoard == 12)[1][0]] in self.whiteMoves:
            print("Black King In Check")
            self.blackKingInCheck = True
        else:
            self.blackKingInCheck = False
            
    def enemyKingInCheckmate(self):
        self.checkMoves = []
        if self.turn == "Black" and self.blackKingInCheck:
            mateFlag = True
            for col in range(size_of_board):
                for row in range(size_of_board):
                    if self.gameBoard[col,row] > 6 and self.gameBoard[col,row] != 0:
                        self.calculateValidMoves([col,row])

                        for move in self.validMoves:
                            if move != [col,row]:
                                tempBoard = copy.copy(self.gameBoard)
                                
                                interim = self.gameBoard[col][row]
                                self.gameBoard[col][row] = 0
                                self.gameBoard[move[0]][move[1]] = interim
                                self.turn = "White"
                                self.calcWhiteAttackMoves()
                                self.turn = "Black"
                                if [np.where(self.gameBoard == 12)[0][0], np.where(self.gameBoard == 12)[1][0]] not in self.whiteMoves:
                                    print(self.gameBoard.transpose())
                                    
                                    
                                    mateFlag = False
                                    self.checkMoves.append(self.gameBoard)
                                self.gameBoard = copy.copy(tempBoard)
                                
            
            if mateFlag:
                print("Checkmate!! White wins!")
                             
                                
        elif self.turn == "White" and self.whiteKingInCheck:
            
            mateFlag = True
            for col in range(size_of_board):
                for row in range(size_of_board):
                    if self.gameBoard[col,row] <= 6 and self.gameBoard[col,row] != 0:
                        self.calculateValidMoves([col,row])

                        for move in self.validMoves:
                            if move != [col,row]:
                                tempBoard = copy.copy(self.gameBoard)
                                
                                interim = self.gameBoard[col][row]
                                self.gameBoard[col][row] = 0
                                self.gameBoard[move[0]][move[1]] = interim
                                
                                self.calcBlackAttackMoves()
                                
                                if [np.where(self.gameBoard == 6)[0][0], np.where(self.gameBoard == 6)[1][0]] not in self.blackMoves:
                                    print(self.gameBoard.transpose())
                                    mateFlag = False
                                
                                    self.checkMoves.append(self.gameBoard)
                                self.gameBoard = copy.copy(tempBoard)
            if mateFlag:
                print("Checkmate!! Black wins!")
            
            
game = Chess()
game.mainloop()
```
